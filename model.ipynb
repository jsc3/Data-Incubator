```python
from os import listdir
import re
from nltk import RegexpTokenizer, FreqDist
import gensim
from gensim.models.doc2vec import TaggedDocument
```

```python
#grab data

docLabels=[]        
docLabels = [f for f in listdir("Data/Total_Data") if f.endswith('.txt')]

data=[]
for doc in docLabels:
    data.append(open('Data/Total_Data/' + doc).read())
```

```python
#data cleaning

regex_list = ["Divide/s[^@]*.*", "Multiply/s[^@]*.*", "Add\s[^@]*.*", "There is[^@]*.*", \
              "[Nn]otes:[^@]*.[^@]*", "[Pp]ost[^@]*.*", "[Rr]eport[^@]*.*", "In this[^@]*.*", \
              "and post to comments.*", "[Cc]ompare[^@]*.*", "to comments.*", "[Cc]omment.*"]

#delete text not related to exercises
data2=[]
for string in data:
    string2 = re.sub("|".join(regex_list), '', string)
    string2 = string2.rstrip()
    data2.append(string2)
    
#delete substrings with first and last name occurences and remove extra lines
#need fixing on regex
data3=[]
sep='DELETE'
for string in data2:
    string2 = re.sub("[A-Z]{1,}[a-z]*\s[a-z]?[']?[A-Z]{1,}[A-Za-z]*[-]?[a-zA-Z]*[:]?\s[\d]+", 'DELETE', string)
    string2 = string2.replace('\n\n', 'DELETE')
    rest = string2.split(sep, 1)[0]
    rest = rest.rstrip()
    final = rest.lower()
    data3.append(final)

#only unique workouts
final_data = list(set(data3))   

#drop bad data
for index, k in enumerate(final_data):
    if len(k) < 6:
        final_data.pop(index)

#copy data
model_data=final_data
        
#If workout has "OR" in it split into two workouts
```

```python
#plot top exercise frequencies

ex_freq = {}
unwanted_chars = "''"
for raw_string in data3:
    string = raw_string.strip(unwanted_chars)
    if string not in ex_freq:
        ex_freq[string] = 0 
    ex_freq[string] += 1
del ex_freq['']
del ex_freq['crossfit']
#most frequent workouts 
ex_freq2 = dict((key, value) for key, value in ex_freq.items() if value > 20)


%matplotlib inline
import matplotlib.pyplot as plt

width=.5
fig = plt.figure(figsize=(75, 20))
ax = fig.add_subplot(111)
plt.tight_layout()

plt.bar(ex_freq2.keys(), ex_freq2.values(), width, color='g')
```

```python
#plot frequencies of top words in workouts -> help to understand types of workouts crossfit offers
#will aid in determining future fitness data needed to make workout generator more well rounded

top_word=[]
top_words=[]
for words in final_data:
    words.split()
    top_word = re.sub("[^\w]", " ",  words).split()
    for i in top_word:
        if not i.isdigit(): 
            top_words.append(i)

word_freq = {}
for word in top_words:
    if word not in word_freq:
        word_freq[word] = 0 
    word_freq[word] += 1

word_freq2 = dict((key, value) for key, value in word_freq.items() if value > 100)

%matplotlib inline
width=.5
fig = plt.figure(figsize=(100, 40))
ax = fig.add_subplot(111)
plt.tight_layout()

plt.bar(word_freq2.keys(), word_freq2.values(), width, color='r')
#the top exercises are more anaerobic, less aerobic (less monotonous workouts = running, rowing, swimming)
#many top exercises are intervals for time, rep-based (weights) or body-weight based (pushups, pullups, situps)
```

```python
#tokenize data for modeling

tokenizer = RegexpTokenizer(r'\w+')
taggeddoc=[]

for index,wod in enumerate(model_data):
    wordslist=[]
    tagslist=[]
    tokens = tokenizer.tokenize(wod)
    
    td = TaggedDocument(gensim.utils.to_unicode(str.encode(' '.join(tokens))).split(), str(index)) 
    taggeddoc.append(td)
    
documents=taggeddoc
#training set
#testting set
```

```python
#instantiate model
model=gensim.models.Doc2Vec(documents, dm=0, dbow_words=1, iter=1, alpha=0.025, min_alpha=0.025, min_count=10)
```

```python
#train model
for epoch in range(100):
    if epoch % 10 == 0:
        print("Training epoch {}".format(epoch))
    model.train(documents, total_examples=model.corpus_count, epochs=model.iter)
    model.alpha -= 0.002
    model.min_alpha = model.alpha
```

```python
#find most similar workout
print(final_data[91])
print(model_data[91])
tokens = model_data[100].split()
new_vector = model.infer_vector(tokens)
sims = model.docvecs.most_similar([new_vector], topn=5)
print(sims)
```
